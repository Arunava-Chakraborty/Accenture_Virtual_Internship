```python
# Import necessary libraries
import pandas as pd

# Load dataset
df = pd.read_csv("/kaggle/input/time-wasters-on-social-media/Time-Wasters on Social Media.csv")

# Check for missing values
print(df.isnull().sum())

# Drop duplicate entries
df.drop_duplicates(inplace=True)

# Convert timestamps to datetime format
df["timestamp"] = pd.to_datetime(df["Total Time Spent"])
df["hour"] = df["timestamp"].dt.hour
df["day"] = df["timestamp"].dt.day_name()

# Check dataset structure after cleaning
print(df.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Countplot for social media platforms usage
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="Platform", order=df["Platform"].value_counts().index, palette="coolwarm")
plt.title("Social Media Platform Usage Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by hour and count usage
hourly_usage = df.groupby("Total Time Spent")["UserID"].count()

# Plot usage trends
plt.figure(figsize=(12, 6))
sns.lineplot(x=hourly_usage.index, y=hourly_usage.values, marker="o", color="blue")
plt.title("Peak Hours of Social Media Usage")
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Users")
plt.grid()
plt.show()
```

```python
# Boxplot to analyze time spent on social media per session
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="Platform", y="Total Time Spent", palette="coolwarm")
plt.title("Time Spent on Social Media per Session")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans

# Sample DataFrame
data = {
    "time_spent": [120, 45, 30, 90, 110],  # Minutes spent on social media
    "hour": ["9:00 PM", "2:30 PM", "7:45 AM", "10:15 PM", "1:00 AM"]
}

df = pd.DataFrame(data)

# Convert 'hour' column to datetime format and extract hour in 24-hour format
df["hour"] = pd.to_datetime(df["hour"], format="%I:%M %p").dt.hour

print(df)  # Check converted values
```

```python
from sklearn.cluster import KMeans
import numpy as np

# Select features for clustering
X = df[["time_spent", "hour"]]

# Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df["user_cluster"] = kmeans.fit_predict(X)

# Visualizing Clusters
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="time_spent", y="hour", hue="user_cluster", palette="coolwarm")
plt.title("User Behavior Clustering")
plt.show()
```
